````python
# train
# ----------------------------
# 0) Windows 穩定性設定（最上面）
# ----------------------------
import os
os.environ["TOKENIZERS_PARALLELISM"] = "false"
os.environ["OMP_NUM_THREADS"] = "1"
os.environ["MKL_NUM_THREADS"] = "1"
os.environ["MKL_SERVICE_FORCE_INTEL"] = "1"
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"
os.environ["UNSLOTH_COMPILE_DISABLE"] = "1"  # ✅ 關 auto compile（Windows 常見救命）

# ----------------------------
# 1) Imports
#    ✅ unsloth 要放在 transformers 前面
# ----------------------------
import gc
import torch
from datasets import load_dataset

from unsloth import FastModel
from unsloth.chat_templates import get_chat_template

from transformers import (
    TrainingArguments,
    Trainer,
    DataCollatorForSeq2Seq,
)

# ----------------------------
# 2) GPU 檢查 + 清理
# ----------------------------
assert torch.cuda.is_available(), "❌ torch 看不到 CUDA，請先確認你是 CUDA 版 PyTorch"
print("torch =", torch.__version__)
print("gpu   =", torch.cuda.get_device_name(0))
print("bf16_supported =", torch.cuda.is_bf16_supported())

gc.collect()
torch.cuda.empty_cache()
torch.cuda.ipc_collect()

free, total = torch.cuda.mem_get_info()
print(f"free GB = {free/1024**3:.2f}, total GB = {total/1024**3:.2f}")

# ----------------------------
# 3) Model / tokenizer
# ----------------------------
max_seq_length = 896
dtype = torch.bfloat16 if torch.cuda.is_bf16_supported() else torch.float16
print("dtype= ", dtype)
model, tokenizer = FastModel.from_pretrained(
    model_name="unsloth/gemma-3-4b-it-bnb-4bit",
    max_seq_length=max_seq_length,
    dtype=dtype,
    load_in_4bit=True,
)

# ✅ 套用 Gemma-3 chat template
tokenizer = get_chat_template(
    tokenizer,
    chat_template="gemma-3",
)

# ✅ 取出底層純文字 tokenizer（很重要）
base_tokenizer = tokenizer.tokenizer if hasattr(tokenizer, "tokenizer") else tokenizer

# ✅ 確保 pad token 存在
if base_tokenizer.pad_token is None:
    base_tokenizer.pad_token = base_tokenizer.eos_token

# ✅ Gemma-3 建議這樣掛 LoRA（純文字訓練）
model = FastModel.get_peft_model(
    model,
    finetune_vision_layers=False,
    finetune_language_layers=True,
    finetune_attention_modules=True,
    finetune_mlp_modules=True,
    r=8,
    lora_alpha=32,
    lora_dropout=0.0,
    bias="none",
    random_state=3407,
)

# ----------------------------
# 4) Load dataset (messages only)
# ----------------------------
file_path = r"C:\Users\CSuser\new_test\jsonl_out\train.jsonl"
dataset = load_dataset("json", data_files={"train": file_path})["train"]
print("columns:", dataset.column_names)
assert "messages" in dataset.column_names, "❌ 你的 jsonl 沒有 messages 欄位"

def is_valid(ex):
    msgs = ex.get("messages", None)
    if not isinstance(msgs, list) or len(msgs) == 0:
        return False
    for m in msgs:
        if not isinstance(m, dict):
            return False
        if "role" not in m or "content" not in m:
            return False
        if not isinstance(m["content"], str):
            return False
        if len(m["content"].strip()) == 0:
            return False
    return True

dataset = dataset.filter(is_valid)
print("valid rows:", len(dataset))

# ----------------------------
# 5) messages -> text
#    ✅ 移除開頭 BOS，避免長度差 1
# ----------------------------
def to_text(ex):
    txt = tokenizer.apply_chat_template(
        ex["messages"],
        tokenize=False,
        add_generation_prompt=False,
    )

    # 移除最前面的 BOS（若存在）
    bos = getattr(base_tokenizer, "bos_token", None)
    if isinstance(txt, str) and bos and txt.startswith(bos):
        txt = txt[len(bos):]

    return {"text": txt}

dataset = dataset.map(to_text)
print("sample text:", dataset[0]["text"][:300])

# -----------------------------
# 6) 檢查是否被截斷
# -----------------------------
max_len = max_seq_length

lengths = [
    len(base_tokenizer(t, add_special_tokens=False).input_ids)
    for t in dataset["text"]
]
n = len(lengths)

trunc_cnt = sum(l > max_len for l in lengths)
print(f"truncated: {trunc_cnt}/{n} = {trunc_cnt/n:.2%}")

# ----------------------------
# 7) Tokenize -> input_ids / attention_mask / labels
#    ✅ 純文字直接用 base_tokenizer，不走 processor
# ----------------------------
def tokenize_one(ex):
    enc = base_tokenizer(
        ex["text"],
        truncation=True,
        max_length=max_seq_length,
        padding=False,
        add_special_tokens=False,
        return_token_type_ids=True,   # ✅ 明確要求回傳
    )

    enc["labels"] = enc["input_ids"].copy()

    # ✅ 如果 tokenizer 沒吐出 token_type_ids，就手動補一條
    if "token_type_ids" not in enc:
        enc["token_type_ids"] = [0] * len(enc["input_ids"])

    return enc

tokenized = dataset.map(
    tokenize_one,
    remove_columns=dataset.column_names,  # 移除 messages / text
)

print("tokenized cols:", tokenized.column_names)
print("sample tokenized lens:",
      len(tokenized[0]["input_ids"]),
      len(tokenized[0]["attention_mask"]),
      len(tokenized[0]["labels"]))

# ----------------------------
# 8) Data collator
# ----------------------------
data_collator = DataCollatorForSeq2Seq(
    tokenizer=base_tokenizer,
    model=None,
    padding=True,
    return_tensors="pt",
)

# ----------------------------
# 9) Trainer
# ----------------------------
args = TrainingArguments(
    output_dir="outputs",
    per_device_train_batch_size=1,
    gradient_accumulation_steps=24,
    max_steps=150,
    learning_rate=5e-5,
    warmup_steps=5,
    fp16=(dtype == torch.float16),
    bf16=(dtype == torch.bfloat16),
    logging_steps=1,
    save_steps=50,
    report_to="none",
    optim="paged_adamw_8bit",
    remove_unused_columns=False,
)

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=tokenized,
    data_collator=data_collator,
)

# ----------------------------
# 10) Train
# ----------------------------
trainer.train()
print("✅ done")
````

````python
# train_new
# ----------------------------
# 0) Windows 穩定性設定（最上面）
# ----------------------------
import os
os.environ["TOKENIZERS_PARALLELISM"] = "false"
os.environ["OMP_NUM_THREADS"] = "1"
os.environ["MKL_NUM_THREADS"] = "1"
os.environ["MKL_SERVICE_FORCE_INTEL"] = "1"
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"
os.environ["UNSLOTH_COMPILE_DISABLE"] = "1"  # ✅ 關 auto compile（Windows 常見救命）

# ----------------------------
# 1) Imports
#    ✅ unsloth 要放在 transformers 前面
# ----------------------------
import gc
import torch
from datasets import load_dataset

from unsloth import FastModel
from unsloth.chat_templates import get_chat_template

from transformers import (
    TrainingArguments,
    Trainer,
    DataCollatorForSeq2Seq,
)

# ----------------------------
# 2) GPU 檢查 + 清理
# ----------------------------
assert torch.cuda.is_available(), "❌ torch 看不到 CUDA，請先確認你是 CUDA 版 PyTorch"
print("torch =", torch.__version__)
print("gpu   =", torch.cuda.get_device_name(0))
print("bf16_supported =", torch.cuda.is_bf16_supported())

gc.collect()
torch.cuda.empty_cache()
torch.cuda.ipc_collect()

free, total = torch.cuda.mem_get_info()
print(f"free GB = {free/1024**3:.2f}, total GB = {total/1024**3:.2f}")

# ----------------------------
# 3) Model / tokenizer
# ----------------------------
max_seq_length = 896
dtype = torch.bfloat16 if torch.cuda.is_bf16_supported() else torch.float16
print("dtype =", dtype)

model, tokenizer = FastModel.from_pretrained(
    model_name="unsloth/gemma-3-4b-it-bnb-4bit",
    max_seq_length=max_seq_length,
    dtype=dtype,
    load_in_4bit=True,
)

# ✅ 套用 Gemma-3 chat template
tokenizer = get_chat_template(
    tokenizer,
    chat_template="gemma-3",
)

# ✅ 取出底層純文字 tokenizer
base_tokenizer = tokenizer.tokenizer if hasattr(tokenizer, "tokenizer") else tokenizer

# ✅ 確保 pad token 存在
if base_tokenizer.pad_token is None:
    base_tokenizer.pad_token = base_tokenizer.eos_token

# ✅ Gemma-3 掛 LoRA（純文字訓練）
model = FastModel.get_peft_model(
    model,
    finetune_vision_layers=False,
    finetune_language_layers=True,
    finetune_attention_modules=True,
    finetune_mlp_modules=True,
    r=8,
    lora_alpha=32,
    lora_dropout=0.0,
    bias="none",
    random_state=3407,
)

# ----------------------------
# 4) Load dataset
# ----------------------------
file_path = r"C:\Users\CSuser\new_test\jsonl_out\train.jsonl"
dataset = load_dataset("json", data_files={"train": file_path})["train"]

print("columns:", dataset.column_names)
assert "messages" in dataset.column_names, "❌ 你的 jsonl 沒有 messages 欄位"

def is_valid(ex):
    msgs = ex.get("messages", None)
    if not isinstance(msgs, list) or len(msgs) == 0:
        return False
    for m in msgs:
        if not isinstance(m, dict):
            return False
        if "role" not in m or "content" not in m:
            return False
        if not isinstance(m["content"], str):
            return False
        if len(m["content"].strip()) == 0:
            return False
    return True

dataset = dataset.filter(is_valid)
print("valid rows:", len(dataset))

# ----------------------------
# 5) 自動切 train / valid
#    ✅ 這裡用 90/10，你也可改成 0.05
# ----------------------------
split_dataset = dataset.train_test_split(test_size=0.1, seed=3407)
train_dataset = split_dataset["train"]
valid_dataset = split_dataset["test"]

print("train rows:", len(train_dataset))
print("valid rows:", len(valid_dataset))

# ----------------------------
# 6) messages -> text
#    ✅ 移除開頭 BOS，避免長度差 1
# ----------------------------
def to_text(ex):
    txt = tokenizer.apply_chat_template(
        ex["messages"],
        tokenize=False,
        add_generation_prompt=False,
    )

    bos = getattr(base_tokenizer, "bos_token", None)
    if isinstance(txt, str) and bos and txt.startswith(bos):
        txt = txt[len(bos):]

    return {"text": txt}

train_dataset = train_dataset.map(to_text)
valid_dataset = valid_dataset.map(to_text)

print("sample text:", train_dataset[0]["text"][:300])

# -----------------------------
# 7) 檢查 train 是否被截斷
# -----------------------------
train_lengths = [
    len(base_tokenizer(t, add_special_tokens=False).input_ids)
    for t in train_dataset["text"]
]
n = len(train_lengths)
trunc_cnt = sum(l > max_seq_length for l in train_lengths)
print(f"train truncated: {trunc_cnt}/{n} = {trunc_cnt/n:.2%}")

# ----------------------------
# 8) Tokenize -> input_ids / attention_mask / labels
# ----------------------------
def tokenize_one(ex):
    enc = base_tokenizer(
        ex["text"],
        truncation=True,
        max_length=max_seq_length,
        padding=False,
        add_special_tokens=False,
        return_token_type_ids=True,
    )

    enc["labels"] = enc["input_ids"].copy()

    if "token_type_ids" not in enc:
        enc["token_type_ids"] = [0] * len(enc["input_ids"])

    return enc

train_tokenized = train_dataset.map(
    tokenize_one,
    remove_columns=train_dataset.column_names,
)

valid_tokenized = valid_dataset.map(
    tokenize_one,
    remove_columns=valid_dataset.column_names,
)

print("train tokenized cols:", train_tokenized.column_names)
print("valid tokenized cols:", valid_tokenized.column_names)
print(
    "sample tokenized lens:",
    len(train_tokenized[0]["input_ids"]),
    len(train_tokenized[0]["attention_mask"]),
    len(train_tokenized[0]["labels"]),
)

# ----------------------------
# 9) Data collator
# ----------------------------
data_collator = DataCollatorForSeq2Seq(
    tokenizer=base_tokenizer,
    model=None,
    padding=True,
    return_tensors="pt",
)

# ----------------------------
# 10) TrainingArguments
# ----------------------------
args = TrainingArguments(
    output_dir="outputs",

    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    gradient_accumulation_steps=24,

    max_steps=150,
    learning_rate=5e-5,
    warmup_steps=5,

    fp16=(dtype == torch.float16),
    bf16=(dtype == torch.bfloat16),

    logging_steps=1,
    disable_tqdm=False,

    # ✅ 每 25 steps 存一次
    save_strategy="steps",
    save_steps=25,

    # ✅ 每 25 steps 驗證一次
    eval_strategy="steps",
    eval_steps=25,

    # ✅ 自動載入最佳 checkpoint
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    greater_is_better=False,

    # ✅ 最多保留幾個 checkpoint
    save_total_limit=3,

    report_to="none",
    optim="paged_adamw_8bit",
    remove_unused_columns=False,
)

# ----------------------------
# 11) Trainer
# ----------------------------
trainer = Trainer(
    model=model,
    args=args,
    train_dataset=train_tokenized,
    eval_dataset=valid_tokenized,
    data_collator=data_collator,
)

# ----------------------------
# 12) Train
# ----------------------------
train_result = trainer.train()

print("✅ done")
print("✅ best checkpoint =", trainer.state.best_model_checkpoint)
print("✅ best metric     =", trainer.state.best_metric)

# ----------------------------
# 13) 存最終 LoRA adapter 與 tokenizer
# ----------------------------
final_save_dir = os.path.join("outputs", "final_adapter")
trainer.save_model(final_save_dir)
base_tokenizer.save_pretrained(final_save_dir)

print(f"✅ final adapter saved to: {final_save_dir}")
````

````python
# valid

import os, json, glob, re
import torch
from datasets import load_dataset
from unsloth import FastModel
from unsloth.chat_templates import get_chat_template
from peft import PeftModel
from tqdm.auto import tqdm

# ----------------------------
# 0) Paths (EDIT THESE)
# ----------------------------
output_dir = r"C:\Users\CSuser\new_test\gemma3\outputs_v6(best)\checkpoint-100"   # 你訓練輸出的資料夾
infer_path = r"C:\Users\CSuser\new_test\jsonl_out\valid.jsonl"
out_path   = r"C:\Users\CSuser\new_test\infer_pred.jsonl"

base_model_name = "unsloth/gemma-3-4b-it-bnb-4bit"
max_seq_length  = 896     # ✅ 跟你 train 的 max_seq_length 一致
MAX_NEW_TOKENS  = 320
print("MAX_NEW_TOKENS =", MAX_NEW_TOKENS)

# ----------------------------
# 1) Find checkpoint
# ----------------------------
adapter_path = output_dir
print("✅ Using adapter path:", adapter_path)

# ----------------------------
# 2) Load base model + adapter (LoRA)
# ----------------------------
assert torch.cuda.is_available(), "❌ CUDA 不可用，請確認你裝的是 CUDA 版 PyTorch"
print("gpu =", torch.cuda.get_device_name(0))
print("bf16_supported =", torch.cuda.is_bf16_supported())

dtype = torch.bfloat16 if torch.cuda.is_bf16_supported() else torch.float16

base_model, tokenizer = FastModel.from_pretrained(
    model_name=base_model_name,
    max_seq_length=max_seq_length,
    dtype=dtype,
    load_in_4bit=True,
)

# ✅ Gemma-3 chat template
tokenizer = get_chat_template(tokenizer, chat_template="gemma-3")

# ✅ LoRA adapter
model = PeftModel.from_pretrained(base_model, adapter_path)

# ✅ inference 模式
FastModel.for_inference(model)
model.eval()

# ----------------------------
# 3) Load inference dataset (messages)
# ----------------------------
def is_valid_messages(ex) -> bool:
    msgs = ex.get("messages", None)
    if not isinstance(msgs, list) or len(msgs) == 0:
        return False
    for m in msgs:
        if not isinstance(m, dict):
            return False
        if "role" not in m or "content" not in m:
            return False
        if not isinstance(m["content"], str):
            return False
    return True

infer_ds = load_dataset("json", data_files={"infer": infer_path})["infer"]
infer_ds = infer_ds.filter(is_valid_messages)
print("infer rows:", len(infer_ds))

# ----------------------------
# 4) Prompt builder + generation (plain text only)
# ----------------------------
SYSTEM_RULE = (
    "你是急診出院病歷摘要助手。"
    "請開頭第一句應優先簡單概述此次入院或住院原因，後續再依序整理相關治療經過純文字段落，且必須忠於輸入資料，不可臆測、不可杜撰、不可補寫輸入中沒有出現的資訊。"
    "禁止輸出任何 XML/HTML/Markdown 標籤或符號（例如 <...>、```、#、*）。"
    "禁止輸出或複製提示詞、規則本身或任何原始欄位名稱。"
    "不要加標題、不要加項目符號、不要加欄位名，只輸出內容本身。"
    "請盡量將治療經過寫得比一般更完整一些，但內容必須完全來自輸入資料。"
    "請主動從各區塊中搜尋可用的治療相關線索並整合成連貫敘述，不可只依賴單一欄位。"
    "若資訊散落在不同段落，請交叉整理並去除重複，寫成自然、正式、簡潔但相對完整的一段"
    "特別注意：<PrimaryDiagnosis> 欄位括號中的補充文字，若涉及病程、處置、檢查、用藥、治療反應、追蹤或醫師說明，應一併納入整理。"
    "若在 <PrimaryDiagnosis>、<SecondaryDiagnosis> 中可用於描述治療經過的資訊不足，請進一步從 <Info>、<Assessment>、<Plan> 或其他相關段落中擷取可用內容加以整合。"
    "禁止只輸出空泛、籠統或資訊量不足的句子，例如'The patient was admitted for further survey and medication. Emergent surgery was arranged.'，應根據原文具體整理為接受了哪些檢查、給予何種治療、以及安排了何種手術或處置。"
    "請優先整理下列資訊：就診原因與主要診斷、主訴、臨床評估、藥物與治療處置、會診建議、出院安排與衛教。"
    "治療經過不可只以籠統方式描述，例如僅寫「進一步檢查」、「給予藥物治療」或「安排手術」。"
    "若原始資料中有提及具體的檢查、處置或手術名稱，必須明確寫出實際執行或安排的內容，例如檢查項目、影像檢查、實驗室檢驗、手術名稱、處置方式或會診建議。"
    "但若輸入資料未明確指出檢查名稱、手術名稱或處置內容，則不可自行補充或推測。"
)

def remove_last_assistant(msgs):
    prompt_msgs = msgs[:]
    for i in range(len(prompt_msgs) - 1, -1, -1):
        if (prompt_msgs[i].get("role") or "").lower() == "assistant":
            return prompt_msgs[:i]
    return prompt_msgs

def build_prompt_text(msgs):
    prompt_msgs = remove_last_assistant(msgs)

    # 覆蓋原本 system
    prompt_msgs = [m for m in prompt_msgs if (m.get("role") or "").lower() != "system"]
    prompt_msgs = [{"role": "system", "content": SYSTEM_RULE}] + prompt_msgs

    # ✅ Gemma-3: 用 chat template 產生 prompt，並加 generation prompt
    return tokenizer.apply_chat_template(
        prompt_msgs,
        tokenize=False,
        add_generation_prompt=True,
    )

TAG_RE = re.compile(r"</?[A-Za-z][^>]*>")
ECHO_MARKERS = ["[輸出規則]", "[任務]", "輸出規則", "禁止輸出", "你只能輸出"]

def postprocess_plain_text(s: str) -> str:
    s = (s or "").strip()
    s = TAG_RE.sub("", s)
    s = s.replace("```", "")
    for mk in ECHO_MARKERS:
        if mk in s:
            s = s.split(mk, 1)[0].strip()
    s = re.sub(r"^\s*(治療經過|Treatment\s*Course)\s*[:：]\s*", "", s, flags=re.IGNORECASE)
    s = s.replace("\r\n", "\n").replace("\r", "\n")
    s = re.sub(r"[ \t]+", " ", s)
    s = re.sub(r"\n{3,}", "\n\n", s).strip()
    return s

def cut_repeated_sentences(text: str, max_repeat: int = 3, sim_threshold: float = 0.88) -> str:
    """
    若連續句子高度相似(>=sim_threshold)重複超過 max_repeat 次，則截斷。
    sim_threshold 調低可抓 ICU/NS/CV ward 交替 loop。
    """
    if not text:
        return text

    import re
    from difflib import SequenceMatcher

    sents = [s.strip() for s in re.split(r'(?<=[。！？!?;；\.])\s+|\n+', text) if s.strip()]
    if not sents:
        return text

    def norm(s: str) -> str:
        s = s.lower()
        s = re.sub(r"\s+", " ", s).strip()
        return s

    out = []
    rep = 0
    prev = None

    for s in sents:
        ns = norm(s)
        if prev is None:
            prev = ns
            out.append(s)
            continue

        sim = SequenceMatcher(None, prev, ns).ratio()
        if sim >= sim_threshold:
            rep += 1
            if rep >= max_repeat:
                break
        else:
            rep = 0
            prev = ns

        out.append(s)

    return " ".join(out).strip()

def has_loop(text: str, window_sent: int = 10, sim_threshold: float = 0.88, min_hits: int = 4) -> bool:
    """
    偵測「變體重複」：最近 window_sent 句中，若出現多句彼此高度相似(>=sim_threshold)的情況達到 min_hits，就判定 loop。
    適合 ICU/NS/CV ward 交替這種句子不完全相同的 loop。
    """
    if not text:
        return False

    import re
    from difflib import SequenceMatcher

    sents = [s.strip() for s in re.split(r'(?<=[。！？!?;；\.])\s+|\n+', text) if s.strip()]
    if len(sents) < 6:
        return False

    tail = sents[-window_sent:]
    def norm(s: str) -> str:
        s = s.lower()
        s = re.sub(r"\s+", " ", s).strip()
        return s

    tail_n = [norm(s) for s in tail]

    hits = 0
    # 看最近句子之間的相似度（任兩句相似就算一次 hit）
    for i in range(len(tail_n)):
        for j in range(i + 1, len(tail_n)):
            if SequenceMatcher(None, tail_n[i], tail_n[j]).ratio() >= sim_threshold:
                hits += 1
                if hits >= min_hits:
                    return True
    return False

def generate_once(prompt_text: str, *, max_new: int, no_repeat=None, rep_pen=None) -> str:
    inputs = tokenizer(text=prompt_text, return_tensors="pt").to("cuda")
    start_len = inputs["input_ids"].shape[-1]

    gen_kwargs = dict(
        max_new_tokens=max_new,
        do_sample=False,
        temperature=0.0,
        pad_token_id=tokenizer.eos_token_id,
        eos_token_id=tokenizer.eos_token_id,
    )
    if no_repeat is not None:
        gen_kwargs["no_repeat_ngram_size"] = no_repeat
    if rep_pen is not None:
        gen_kwargs["repetition_penalty"] = rep_pen

    with torch.no_grad():
        outputs = model.generate(**inputs, **gen_kwargs)

    gen_ids = outputs[0][start_len:]
    text = tokenizer.decode(gen_ids, skip_special_tokens=True).strip()
    return postprocess_plain_text(text)

def generate_answer(prompt_text: str) -> str:
    text = generate_once(prompt_text, max_new=MAX_NEW_TOKENS)

    if has_loop(text):  # ✅ 現在會抓到變體 loop
        text = generate_once(prompt_text, max_new=MAX_NEW_TOKENS, no_repeat=4, rep_pen=1.06)

    # ✅ 門檻放鬆到 0.88，比較抓得到 ICU/NS/CV 交替
    text = cut_repeated_sentences(text, max_repeat=3, sim_threshold=0.88)
    return text

# ----------------------------
# 5) Run inference + save jsonl
# ----------------------------
with open(out_path, "w", encoding="utf-8") as f:
    for ex in tqdm(infer_ds, total=len(infer_ds), desc="Valid inference"):
        msgs = ex["messages"]
        prompt_text = build_prompt_text(msgs)
        pred = generate_answer(prompt_text)

        new_ex = dict(ex)
        new_ex["pred"] = pred
        f.write(json.dumps(new_ex, ensure_ascii=False) + "\n")

print("✅ saved:", out_path)
````

````python
#評估指標_new
import json, re, math
from collections import Counter
from pathlib import Path

# ----------------------------
# Settings (edit)
# ----------------------------
PRED_JSONL = r"C:\Users\CSuser\new_test\infer_pred.jsonl"  # 你的推論輸出，含 messages + pred
MAX_CHARS = 800     # 格式/長度規則：最長（你可依院內習慣調）
MIN_CHARS = 40
# ----------------------------
# Utilities
# ----------------------------
SENT_SPLIT_RE = re.compile(r"[。！？!?;\n]+")
TAG_RE = re.compile(r"<[^>]+>")
MD_RE = re.compile(r"(^|\n)\s*([#\-\*\•]|```)", re.MULTILINE)

def normalize_text(s: str) -> str:
    s = s or ""
    s = s.replace("\r\n", "\n").replace("\r", "\n")
    s = s.replace("_x000D_", "\n")
    s = re.sub(r"[ \t]+", " ", s)
    s = re.sub(r"\n{3,}", "\n\n", s)
    return s.strip()

def split_sentences(s: str):
    s = normalize_text(s)
    parts = [p.strip() for p in SENT_SPLIT_RE.split(s) if p.strip()]
    return parts

def get_last_role_content(messages, role):
    for m in reversed(messages):
        if m.get("role") == role:
            return m.get("content", "")
    return ""

# ----------------------------
# Keyword-based claim/entity extraction
# (你可以後面再加字典；先用這版就能跑出數值)
# ----------------------------
MED_PATTERNS = [
    r"\b(mannitol|diamox|acetazolamide|simbrinza|duotrav|cefazolin|vancomycin|gentamicin|amoxicillin|augmentin|ibuprofen|acetaminophen|paracetamol|morphine|tramadol|ketorolac)\b",
    r"\b(IV|IM|PO|SC|STAT|BID|TID|QID|q\d+h)\b",  # 給藥途徑/頻率（常見縮寫）
    r"\b(amp|tab|cap|mg|g|ml|mL|IU)\b",
]

TEST_PATTERNS = [
    r"\b(CT|MRI|X[- ]?ray|CXR|ECG|EKG|US|ultrasound)\b",
    r"\b(lab|blood test|urinalysis|UA|CBC|CRP|BUN|creatinine|troponin)\b",
    r"\b(抽血|驗血|尿檢|檢驗|心電圖|超音波|X光|電腦斷層|核磁)\b",
]

PROC_PATTERNS = [
    r"\b(consult|consulted|consultation|OPD|follow[- ]?up|discharge|admit|admission)\b",
    r"\b(IVF|infusion|transfusion|suturing|suture|biopsy|operation|surgery)\b",
    r"\b(眼科|會診|出院|住院|追蹤|轉診|手術|縫合|輸液|輸血)\b",
]

def extract_terms(text: str, patterns):
    text_n = normalize_text(text).lower()
    found = []
    for pat in patterns:
        for m in re.finditer(pat, text_n, flags=re.IGNORECASE):
            found.append(m.group(0).lower())
    # 去重但保留次序
    seen = set()
    uniq = []
    for t in found:
        if t not in seen:
            seen.add(t)
            uniq.append(t)
    return uniq

def extract_entities(pred: str):
    meds = extract_terms(pred, MED_PATTERNS)
    tests = extract_terms(pred, TEST_PATTERNS)
    procs = extract_terms(pred, PROC_PATTERNS)
    return meds, tests, procs

def term_supported(term: str, input_text: str) -> bool:
    # 在 input 中找 term（簡單包含判斷；可再強化同義詞）
    return term.lower() in normalize_text(input_text).lower()

# ----------------------------
# Metric 1: Unsupported claim rate (sentence-level)
# ----------------------------
def unsupported_claim_rate(pred: str, input_text: str):
    sents = split_sentences(pred)
    if not sents:
        return 0.0, 0, 0

    claim_sents = 0
    unsupported = 0

    for sent in sents:
        meds, tests, procs = extract_entities(sent)
        terms = meds + tests + procs
        if not terms:
            continue  # 沒有具體醫療事件詞，就不當 claim
        claim_sents += 1
        # 句子中的任何一個 term 若都找不到依據 -> 算 unsupported（保守）
        supported_any = any(term_supported(t, input_text) for t in terms)
        if not supported_any:
            unsupported += 1

    rate = (unsupported / claim_sents) if claim_sents > 0 else 0.0
    return rate, unsupported, claim_sents

# ----------------------------
# Metric 2: Entity/Medication/Procedure Precision (entity-level)
# ----------------------------
def entity_precision(pred: str, input_text: str):
    meds, tests, procs = extract_entities(pred)
    all_terms = meds + tests + procs
    if not all_terms:
        return 1.0, 0, 0, {"meds": (0,0), "tests": (0,0), "procs": (0,0)}  # 沒有實體就視為不亂補

    def count_supported(terms):
        sup = sum(1 for t in terms if term_supported(t, input_text))
        return sup, len(terms)

    sup_m, tot_m = count_supported(meds)
    sup_t, tot_t = count_supported(tests)
    sup_p, tot_p = count_supported(procs)

    sup_all = sup_m + sup_t + sup_p
    tot_all = tot_m + tot_t + tot_p
    prec = sup_all / tot_all if tot_all else 1.0

    detail = {
        "meds": (sup_m, tot_m),
        "tests": (sup_t, tot_t),
        "procs": (sup_p, tot_p),
    }
    return prec, sup_all, tot_all, detail

# ----------------------------
# Metric 3: ROUGE-L (F1) - simple implementation
# ----------------------------
def lcs_length(a, b):
    # a,b: list of tokens
    # DP O(n*m) - ok for short texts
    n, m = len(a), len(b)
    dp = [0] * (m + 1)
    for i in range(1, n + 1):
        prev = 0
        for j in range(1, m + 1):
            tmp = dp[j]
            if a[i-1] == b[j-1]:
                dp[j] = prev + 1
            else:
                dp[j] = max(dp[j], dp[j-1])
            prev = tmp
    return dp[m]

def rouge_l_f1(pred: str, ref: str):
    pred = normalize_text(pred)
    ref = normalize_text(ref)
    if not pred or not ref:
        return 0.0
    # 用簡單 token：英文用空白切，中文因為沒分詞就用字元序列
    # 混合語言時，這樣雖粗但能跑
    def to_seq(s):
        if re.search(r"[\u4e00-\u9fff]", s):  # 有中文 -> 用字元序列（去空白）
            return [ch for ch in s if ch.strip()]
        else:
            return s.lower().split()

    p_seq = to_seq(pred)
    r_seq = to_seq(ref)
    lcs = lcs_length(p_seq, r_seq)
    prec = lcs / len(p_seq) if p_seq else 0.0
    rec = lcs / len(r_seq) if r_seq else 0.0
    if prec + rec == 0:
        return 0.0
    return 2 * prec * rec / (prec + rec)

# ----------------------------
# Metric 4: Format/length rules
# ----------------------------
def format_length_checks(pred: str):
    s = normalize_text(pred)
    ok_len = (MIN_CHARS <= len(s) <= MAX_CHARS)
    has_tag = bool(TAG_RE.search(s))
    has_md = bool(MD_RE.search(s))
    # 你要「純文字治療經過」：tag / markdown 任一出現就算 fail
    ok_format = (not has_tag) and (not has_md)
    return {
        "len": len(s),
        "ok_len": ok_len,
        "has_tag": has_tag,
        "has_md": has_md,
        "ok_format": ok_format,
        "ok_all": ok_len and ok_format,
    }

# ----------------------------
# Main
# ----------------------------
def main():
    path = Path(PRED_JSONL)
    assert path.exists(), f"File not found: {path}"

    n = 0
    u_rates = []
    u_cnt = 0
    c_cnt = 0

    precs = []
    sup_ent = 0
    tot_ent = 0
    detail_sum = {"meds": [0,0], "tests": [0,0], "procs": [0,0]}

    rougeLs = []
    fmt_ok = 0
    fmt_len_ok = 0
    fmt_tag = 0
    fmt_md = 0

    with path.open("r", encoding="utf-8") as f:
        for line in f:
            if not line.strip():
                continue
            ex = json.loads(line)
            messages = ex.get("messages", [])
            pred = ex.get("pred", "")

            # input: last user
            inp = get_last_role_content(messages, "user")
            # reference: last assistant
            ref = get_last_role_content(messages, "assistant")

            inp = normalize_text(inp)
            pred = normalize_text(pred)
            ref = normalize_text(ref)

            if not pred:
                continue

            n += 1

            # 1) Unsupported claim rate
            rate, u, c = unsupported_claim_rate(pred, inp)
            u_rates.append(rate)
            u_cnt += u
            c_cnt += c

            # 2) Entity precision
            prec, sup, tot, detail = entity_precision(pred, inp)
            precs.append(prec)
            sup_ent += sup
            tot_ent += tot
            for k in detail_sum:
                detail_sum[k][0] += detail[k][0]
                detail_sum[k][1] += detail[k][1]

            # 3) ROUGE-L
            if ref:
                rougeLs.append(rouge_l_f1(pred, ref))

            # 4) Format/length
            fm = format_length_checks(pred)
            if fm["ok_len"]:
                fmt_len_ok += 1
            if fm["has_tag"]:
                fmt_tag += 1
            if fm["has_md"]:
                fmt_md += 1
            if fm["ok_all"]:
                fmt_ok += 1

    if n == 0:
        print("No valid rows.")
        return

    avg_u = sum(u_rates) / len(u_rates) if u_rates else 0.0
    avg_prec = sum(precs) / len(precs) if precs else 1.0
    micro_prec = (sup_ent / tot_ent) if tot_ent else 1.0
    avg_rougeL = sum(rougeLs) / len(rougeLs) if rougeLs else 0.0

    print("==== Metrics ====")
    print(f"Samples: {n}")
    print(f"[Unsupported claim rate] avg={avg_u:.4f}  (unsupported={u_cnt}, claim_sents={c_cnt})")
    print(f"[Entity precision] macro(avg per-sample)={avg_prec:.4f}  micro(overall)={micro_prec:.4f}  (supported={sup_ent}, total={tot_ent})")
    print(f"   meds:  {detail_sum['meds'][0]}/{detail_sum['meds'][1]}  tests: {detail_sum['tests'][0]}/{detail_sum['tests'][1]}  procs: {detail_sum['procs'][0]}/{detail_sum['procs'][1]}")
    print(f"[ROUGE-L F1] avg={avg_rougeL:.4f}  (computed on {len(rougeLs)} refs)")
    print("---- Format/Length ----")
    print(f"ok_all={fmt_ok}/{n}  ok_len={fmt_len_ok}/{n}  has_tag={fmt_tag}/{n}  has_md={fmt_md}/{n}")

if __name__ == "__main__":
    main()
````

````python
#臆測率/合規率
import json, re
from collections import Counter
import pandas as pd

# =========================
# 0) 你的檔案路徑（改這裡）
# =========================
JSONL_PATH = r"C:\Users\CSuser\new_test\infer_pred.jsonl"

# =========================
# 1) 小工具：把 messages 裡的「輸入內容」串起來
# =========================
def build_source_text(obj):
    """
    來源文字 = messages 裡的 system + user content（不含 assistant）
    """
    msgs = obj.get("messages", [])
    parts = []
    for m in msgs:
        role = (m.get("role") or "").lower()
        if role in ("system", "user"):
            parts.append(m.get("content", "") or "")
    return "\n".join(parts)

def get_reference_text(obj):
    """
    參考答案 = messages 中 assistant 的 content（如果有的話）
    """
    msgs = obj.get("messages", [])
    for m in msgs:
        if (m.get("role") or "").lower() == "assistant":
            return m.get("content", "") or ""
    return ""

def normalize(s: str) -> str:
    s = s or ""
    s = s.replace("_x000D_", "\n")
    s = re.sub(r"\s+", " ", s).strip()
    return s

# =========================
# 2) 你要抓的「容易臆測」關鍵詞/片語（可自行加）
#    規則：pred 出現 AND source 沒出現 => 視為臆測命中
# =========================
HALLUCINATION_PATTERNS = [
    # Imaging / tests (EN)
    r"\bbrain ct\b", r"\bhead ct\b", r"\bct\b", r"\bcxr\b", r"\bx[- ]?ray\b",
    r"\bchest x[- ]?ray\b", r"\bmri\b", r"\bultrasound\b", r"\bus\b",
    r"\bekg\b", r"\becg\b",
    # Typical invented findings (EN)
    r"\bno ich\b", r"\bich\b", r"\bno st[- ]?t\b", r"\bst[- ]?t\b",
    r"\bno significant finding\b", r"\bno acute\b",
    r"\bsinus tachycardia\b", r"\btachycardia\b",
    # Imaging / tests (ZH)
    r"腦部ct", r"頭部ct", r"胸部x光", r"胸部x-ray", r"腦出血", r"顱內出血", r"無出血",
    r"心電圖", r"胸片", r"x光", r"超音波", r"核磁共振",
]

halluc_re = [re.compile(p, re.IGNORECASE) for p in HALLUCINATION_PATTERNS]

# =========================
# 3) 合規檢查：抓「不是治療經過」常見標籤/段落抄貼
# =========================
NON_TREATMENT_MARKERS = [
    "<primarydiagnosis>", "<secondarydiagnosis>", "<chiefcomplaint>",
    "<historyofpresentillness>", "<consultationnotes>", "<labreports>",
    "chief complaint", "history of present illness", "primarydiagnosis",
    "secondarydiagnosis", "consultationnotes", "labreports",
    "病史/理學檢查", "理學檢查", "實驗數據", "評估(assessment", "建議(plan",
    "assessment/impression", "plan/recommendation", "mse:", "cc：", "pi：",
    "####",
]

def check_compliance(pred: str):
    """
    回傳 (is_compliant, reasons:list)
    """
    pred_n = pred or ""
    low = pred_n.lower()
    reasons = []

    # 3-1) 出現明顯「非治療經過」段落標籤
    for mk in NON_TREATMENT_MARKERS:
        if mk.lower() in low:
            reasons.append(f"contains_marker:{mk}")
            break

    # 3-2) 太長也常代表把病史整段貼出來（可調）
    if len(pred_n) > 800:
        reasons.append("too_long>800chars")

    # 3-3) 如果 pred 跟 source 重疊度過高，可能是直接抄輸入（簡單偵測）
    # 用 5-gram token overlap 估一個 rough 指標
    def tokens(s):
        s = re.sub(r"[^0-9A-Za-z\u4e00-\u9fff]+", " ", s)
        return [t for t in s.split() if t]
    return (len(reasons) == 0, reasons)

# =========================
# 4) 讀 JSONL + 計算
# =========================
rows = []
with open(JSONL_PATH, "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        line = line.strip()
        if not line:
            continue
        obj = json.loads(line)

        source = normalize(build_source_text(obj))
        ref = normalize(get_reference_text(obj))
        pred = normalize(obj.get("pred", ""))

        # hallucination hits
        hits = []
        for cre, pat in zip(halluc_re, HALLUCINATION_PATTERNS):
            if cre.search(pred) and (not cre.search(source)):
                hits.append(pat)

        # compliance
        compliant, reasons = check_compliance(pred)

        rows.append({
            "idx": i,
            "pred_len": len(pred),
            "halluc_hit_count": len(hits),
            "halluc_hits": "; ".join(hits),
            "compliant": compliant,
            "compliance_reasons": "; ".join(reasons),
            "reference": ref,
            "pred": pred,
        })

df = pd.DataFrame(rows)

# =========================
# 5) Summary
# =========================
n = len(df)
halluc_any = (df["halluc_hit_count"] > 0).sum()
halluc_rate = halluc_any / n if n else 0.0
avg_hits = df["halluc_hit_count"].mean() if n else 0.0
comp_rate = df["compliant"].mean() if n else 0.0

print("==== Summary ====")
print(f"Total samples: {n}")
print(f"Hallucination rate (any hit): {halluc_rate:.3f} ({halluc_any}/{n})")
print(f"Avg hallucination hits/sample: {avg_hits:.3f}")
print(f"Compliance rate: {comp_rate:.3f}")

# Top hallucinated keywords
counter = Counter()
for s in df["halluc_hits"].fillna(""):
    if s.strip():
        for item in s.split(";"):
            item = item.strip()
            if item:
                counter[item] += 1

print("\n==== Top hallucinated patterns ====")
for k, v in counter.most_common(20):
    print(f"{v:4d}  {k}")

# Worst cases
TOP_N = 10
worst = df.sort_values(["halluc_hit_count", "pred_len"], ascending=False).head(TOP_N)

print(f"\n==== Worst {TOP_N} samples ====")
for _, r in worst.iterrows():
    print("\n------------------------------")
    print(f"idx={r['idx']}  hits={r['halluc_hit_count']}  compliant={r['compliant']}")
    if r["halluc_hits"]:
        print("hits:", r["halluc_hits"])
    if r["compliance_reasons"]:
        print("compliance_reasons:", r["compliance_reasons"])
    print("pred:", (r["pred"][:300] + ("..." if len(r["pred"]) > 300 else "")))

# Save report
out_csv = re.sub(r"\.jsonl$", "", JSONL_PATH) + "_eval_report.csv"
df.to_csv(out_csv, index=False, encoding="utf-8-sig")
print("\n✅ Saved report:", out_csv)
````

````python
#評估指標
import json, re, math
import jieba
import torch
import nltk
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from nltk.translate.meteor_score import meteor_score
from rouge_score import rouge_scorer

# 第一次跑需要
nltk.download("wordnet")
nltk.download("omw-1.4")

infer_pred_path = r"C:\Users\CSuser\new_test\infer_pred.jsonl"

# -----------------------
# Text utils
# -----------------------
def remove_punctuation(text):
    # 保留中文/英數/底線/空白
    return re.sub(r"[^\w\s\u4e00-\u9fff]", "", text)

def tokenize_chinese(text):
    return " ".join(jieba.cut(text))

def extract_ref_from_messages(messages):
    # 取最後一個 assistant content 當 reference
    assistants = [m.get("content","") for m in messages if m.get("role") == "assistant"]
    return assistants[-1].strip() if assistants else ""

# -----------------------
# Metrics
# -----------------------
smoothie = SmoothingFunction().method4
rouge = rouge_scorer.RougeScorer(["rouge1","rouge2","rougeL"], use_stemmer=False)

def compute_bleu(reference, candidate):
    ref_tok  = tokenize_chinese(remove_punctuation(reference)).split()
    cand_tok = tokenize_chinese(remove_punctuation(candidate)).split()
    if len(ref_tok) == 0 or len(cand_tok) == 0:
        return 0.0
    return sentence_bleu([ref_tok], cand_tok, smoothing_function=smoothie)

def compute_rouge(reference, candidate):
    return rouge.score(reference, candidate)

def compute_meteor(reference, candidate):
    ref_tok  = tokenize_chinese(remove_punctuation(reference)).split()
    cand_tok = tokenize_chinese(remove_punctuation(candidate)).split()
    if len(ref_tok) == 0 or len(cand_tok) == 0:
        return 0.0
    return meteor_score([ref_tok], cand_tok)

# ⚠️ Perplexity 需要你目前環境已經有 model + tokenizer（推論時那組）
# 這裡用 pred 本身去算 PPL（不是最有意義的評估，但你要的指標可以算）
def compute_perplexity(model, tokenizer, text, max_len=1024):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=max_len).to("cuda")
    with torch.no_grad():
        outputs = model(**inputs, labels=inputs["input_ids"])
        loss = outputs.loss
    return float(math.exp(loss.item()))

# -----------------------
# Load (ref, pred) pairs
# -----------------------
refs, preds = [], []
with open(infer_pred_path, "r", encoding="utf-8") as f:
    for line in f:
        obj = json.loads(line)
        pred = obj.get("pred", "").strip()
        ref = ""

        if "messages" in obj:
            ref = extract_ref_from_messages(obj["messages"])

        if ref and pred:
            refs.append(ref)
            preds.append(pred)

print("loaded pairs:", len(preds))

# -----------------------
# Compute metrics
# -----------------------
bleu_scores  = [compute_bleu(r, p) for r, p in zip(refs, preds)]
rouge_scores = [compute_rouge(r, p) for r, p in zip(refs, preds)]
meteor_scores = [compute_meteor(r, p) for r, p in zip(refs, preds)]

avg_bleu = sum(bleu_scores) / len(bleu_scores)
avg_rouge = {k: sum(s[k].fmeasure for s in rouge_scores) / len(rouge_scores) for k in rouge_scores[0]}
avg_meteor = sum(meteor_scores) / len(meteor_scores)

# Perplexity（需要 model/tokenizer 已存在）
perplexity_scores = [compute_perplexity(model, tokenizer, p) for p in preds]
avg_perplexity = sum(perplexity_scores) / len(perplexity_scores)

print(f"Average BLEU Score: {avg_bleu}")
print(f"Average ROUGE Scores: {avg_rouge}")
print(f"Average METEOR Score: {avg_meteor}")
print(f"Average Perplexity: {avg_perplexity}")
````
